```python
import os
import pandas as pd
#file_path = 'C:/Users/marcb/Documents/pflotran-darcy-velocity-in-reaction-sandbox/src/pflotran/richards_common.F90'



# Function to read content from the provided Fortran file and generate next-token pairs

def generate_next_token_words(text, input_length=20, target_length=10):
    words = text.split()
    input_target_pairs = []
    for i in range(len(words) - input_length - target_length + 1):
        input_seq = ' '.join(words[i:i + input_length])
        target_seq = ' '.join(words[i + input_length:i + input_length + target_length])
        input_target_pairs.append((input_seq, target_seq))
    return input_target_pairs

def generate_next_token_letters(file_path):
    with open(file_path, 'r') as file:
        content = file.read()  # Reading only a limited amount of content
    # Generate input-target pairs for strings of words
    input_target_pairs = generate_next_token_words(content)
    # Generate input-target pairs character-by-character
    #input_target_pairs = [(content[i:i+5], content[i+5:i+7]) for i in range(len(content) - 5)]
    return input_target_pairs

def create_qa_pair(file_path):
    """
    Create a question-answer pair for a given module file.
    The question asks for the source code of the module, and the answer is the file's content.

    :param file_path: Path to the module file.
    :return: A tuple containing the question and the answer.
    """
    # Extracting the module name from the file path
    module_name = os.path.basename(file_path).split('.')[0]

    # Formulating the question
    question = f"What is the source code for the '{module_name}' in PFLOTRAN?"

    # Reading the file content to use as the answer
    try:
        with open(file_path, 'r') as file:
            answer = file.read()
    except Exception as e:
        answer = f"Error reading file: {e}"

    return question, answer

# def bobobob(file_path):
#     with open(file_path, 'r') as file:
#         content = file.read().split()
#     count=0
#     filtered_list = [item for item in content if item.strip()]
#     # Generate input-target pairs character-by-character
#     input_target_pairs = [(filtered_list[i:i+10], filtered_list[i+10:i+20][) for i in range(len(filtered_list) - 2)]
#     #input_target_pairs = pd.DataFrame(input_target_pairs)
#     print (input_target_pairs)
#     return input_target_pairs

# Generating next-token pairs from the provided .F90 file
base_dir = 'C:/Users/marcb/Documents/pflotran-darcy-velocity-in-reaction-sandbox/src/pflotran/'
files = os.listdir(base_dir)
input_target_pairs=[]
count=0
for i in files:
    if ".F90" in i:
        print (count)
        count = count+1
        try:
            file_path = base_dir+i
            input_target_pairs.append(create_qa_pair(file_path))
        except:
            print("Error with module tokenization")

# Displaying the number of pairs generated and a few example pairs
num_pairs = len(input_target_pairs)
example_pairs = input_target_pairs[:5]  # Displaying first 5 pairs for example

num_pairs, example_pairs
```

```python
import numpy as np
full_pairs = np.vstack(input_target_pairs)
full_pairs[0]
```
